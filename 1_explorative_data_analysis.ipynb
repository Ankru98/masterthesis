As we can see from data.head(), the first rows of data are NaN. Therefore we skip the first year and start the EDA at 1987. 

```python
# values from 01-01-1987 until 01-12-20222
price_data = data[12:]
```

```python
price_data.tail()
```

```python
# Calculate the number of non-null values in each column
num_non_null = price_data.count()

# Find the columns with less than 432 non-null values
less_than_432 = num_non_null[num_non_null < 432]
more_than_432 = num_non_null[num_non_null >= 432]

# Print the column names
print(list(less_than_432.index))
```

Column with less than 432 non-null values (36 years):

1. Energy
2. Real Estate
3. Food Bev
4. Cons Prd & Svs
5. Retail
6. Travel & Leisure

```python
# Calculate the number of non-null values in each column
num_non_null = price_data.count()

# Find the columns with less than 432 non-null values
less_than_300 = num_non_null[num_non_null < 300]

# Print the column names
print(list(less_than_300.index))
```

Values with less than 300 non-null values (25 years):

1. Energy
2. Real Estate
3. Food Bev
4. Cons Prd & Svs

So either we could work with a full data set of 35 years including the STOXX Europe 600 as a benchmark and 13 out of 19 sector indices  OR a full data set of 25 years including the STOXX Europe 600 as a benchmark and 15 out of 19 sector indices.

After a discussion with my professor we decised to work with the data set of 36 years.

```python
# save table as csv file
# agreement with Prof. Dr. Markus Haas on 36 years

data_filter = price_data.loc['1987-01-01':'2022-01-01']
data_filter = data_filter[list(more_than_432.index)]
data_filter.head()
```

```python
data_filter.to_csv('data/price_data_filtered.csv')
```
