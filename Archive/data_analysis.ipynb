You can use in further steps the info command (price_data.info()) to receive necessary information and conclude about the missing values and how to handle them.
Downloaded interval: 27.02.1998 - 27.02.2023 (monthly).

```python
# apply command to show only columns with at least 80% values and name it price_data_80
price_data_80 = price_data.loc[:, pd.notnull(price_data).sum()>len(price_data)*0.8]
```

The Info-function shows once more that there are some missing values in the data. Therefore, we exclude them and restrict ourselves to end our research in 2022 as well.

```python
price_data_final = price_data_80.iloc[71:419]
price_data_final.info() # proves that every column has 348 values without any missing values
```
